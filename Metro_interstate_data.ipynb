```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Metro_Interstate_Traffic_Volume.csv')
# pd.set_option('display.max_rows',20)
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()

# Fit and transform the 'Category' column
df['weather_main'] = label_encoder.fit_transform(df['weather_main'])

# Display the DataFrame
df
```

```python
df1=df[['date_time', 'traffic_volume']]
```

```python
df1['date_time'] = pd.to_datetime(df1['date_time'])
df1.set_index('date_time', inplace=True)

# Display the DataFrame
print(df1)
```

```python
df1.fillna(0)
```

```python
df1.index
```

```python
df_daily = df1.resample('12H').mean()
```

```python
df_daily.fillna(df_daily.median(), inplace=True)
```

```python
df_daily.shape
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_daily['traffic_volume'], linestyle='-')
# plt.xlim(df1[start_date], df1[end_date])
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df_daily['traffic_volume'], model='additive', period=12)

# Plot the original, trend, seasonal, and residual components
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(df_daily['traffic_volume'], label='Original')
plt.legend()
plt.title('Original Time Series Data')

plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Trend')
plt.legend()
plt.title('Trend Component')

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Seasonal')
plt.legend()
plt.title('Seasonal Component')

plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residual')
plt.legend()
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plt.figure(figsize=(12, 4))
plot_acf(df_daily['traffic_volume'], title='Autocorrelation Function (ACF)')
plt.show()

# Plot Partial Autocorrelation Function (PACF)
plt.figure(figsize=(12, 4))
plot_pacf(df_daily['traffic_volume'], title='Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
# Dickey-Fuller test
result = adfuller(df_daily['traffic_volume'])

# Print the test statistic and p-value
print(f'Test Statistic: {result[0]}')
print(f'p-value: {result[1]}')

# Print critical values
print('Critical Values:')
for key, value in result[4].items():
    print(f'   {key}: {value}')
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Fit ARIMA model
order = (1, 1, 0)  # Example order, replace with the values you derived from ACF and PACF plots
arima_model = ARIMA(df_daily, order=order)
arima_result = arima_model.fit()

# Generate predictions
forecast_steps =60
# Number of steps to forecast
forecast = arima_result.get_forecast(steps=forecast_steps)
```

```python
forecast.row_labels
```

```python
df_daily['traffic_volume']
```

```python
forecast.predicted_mean
```

```python
forecast.conf_int().iloc[:, 0]
```

```python
# Plot the original data and forecast
plt.figure(figsize=(12, 6))
plt.plot(df_daily.index, df_daily, label='Original Data', linestyle='-')
plt.plot(forecast.row_labels, forecast.predicted_mean, label='Forecast', color='red', linestyle='--')
plt.fill_between(forecast.row_labels, forecast.conf_int().iloc[:, 0], forecast.conf_int().iloc[:, 1], color='red', alpha=0.2)
plt.title('ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
